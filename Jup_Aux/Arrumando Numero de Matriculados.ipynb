```python
df = pd.read_csv('matriculas.csv')

df.columns
cols=[i for i in df.columns if i not in ["Municipios"]]
for col in cols:
    df[col] = df[col].replace(',','',regex=True).astype(float)

df.loc[:,'Total Matriculados'] = df.sum(axis = 1) 

cols=[i for i in df.columns if i not in ["Municipios"]]
for col in cols:
    df[col] = df[col].round()
    
matriculados = df[['Municipios', 'Total Matriculados']]
matriculados

export_final = df.to_csv (r'matriculados.csv', index = None, header=True)
export_final
```

```python
transferencias = pd.read_csv('transferencias_para_municipios.csv')
del transferencias['Ano']
del transferencias['Transferência']
transferencias
```

```python
compacto = pd.DataFrame(columns = ['Municipio','FUNDEB', 'Total Matriculados'])
compacto['Municipio'] = transferencias['Município']
compacto['FUNDEB'] = transferencias['Valor Consolidado']
compacto['Total Matriculados'] = matriculados['Total Matriculados']
compacto['Custo Aluno'] = (compacto['FUNDEB']/compacto['Total Matriculados']).round(2)
compacto
```

```python
base = pd.read_csv('EduBas_Munic.csv')
base
df_join = base.set_index('NomeMun').join(compacto.set_index('Municipio'), how='left')
df_join
```
