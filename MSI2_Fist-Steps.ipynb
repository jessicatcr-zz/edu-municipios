```python
# -*- coding: utf 8

from matplotlib import pyplot as plt

import pandas as pd
import numpy as np
```

```python
dfMunicipio = pd.read_csv('Base_Munic_2017_Pop.csv')
dfMunicipio_MG = dfMunicipio.loc[dfMunicipio['UF']=='MG']
#dfMunicipio
dfMunicipio_MG
```

```python
#Total de municipios brasileiros
print('BRA:', dfMunicipio['UF'].count())

#Total de municipios mineiros
print('MG', dfMunicipio_MG['UF'].count())
```

```python
#Brasileiros
classPop_1 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="1 - Até 5000").count()
classPop_2 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="2 - 5001 até 10000").count()
classPop_3 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="3 - 10001 até 20000").count()
classPop_4 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="4 - 20001 até 50000").count()
classPop_5 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="5 - 50001 até 100000").count()
classPop_6 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="6 - 100001 até 500000").count()
classPop_7 = dfMunicipio.where(dfMunicipio["CLASSE POP"]=="7 - Maior que 500000").count()
total_ate20k = classPop_1["CLASSE POP"] + classPop_2["CLASSE POP"] + classPop_3["CLASSE POP"]
print('BRA:', total_ate20k)

#Mineiros
classPop_MG_1 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="1 - Até 5000").count()
classPop_MG_2 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="2 - 5001 até 10000").count()
classPop_MG_3 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="3 - 10001 até 20000").count()
classPop_MG_4 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="4 - 20001 até 50000").count()
classPop_MG_5 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="5 - 50001 até 100000").count()
classPop_MG_6 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="6 - 100001 até 500000").count()
classPop_MG_7 = dfMunicipio_MG.where(dfMunicipio_MG["CLASSE POP"]=="7 - Maior que 500000").count()
total_MG_ate20k = classPop_MG_1["CLASSE POP"] + classPop_MG_2["CLASSE POP"] + classPop_MG_3["CLASSE POP"]
print('MG:', total_MG_ate20k)
```

```python
classLabel = 'Até 5.000', '5.001 até 10.000', '10.001 até 20.000', '20.001 até 50.000', '50.001 até 100.000', '100.001 até 500.000', 'Maior que 500.000'
classSizes = [classPop_1["CLASSE POP"], classPop_2["CLASSE POP"], classPop_3["CLASSE POP"],
              classPop_4["CLASSE POP"], classPop_5["CLASSE POP"], classPop_6["CLASSE POP"], classPop_7["CLASSE POP"]]

classMGSizes = [classPop_MG_1["CLASSE POP"], classPop_MG_2["CLASSE POP"], classPop_MG_3["CLASSE POP"],
              classPop_MG_4["CLASSE POP"], classPop_MG_5["CLASSE POP"], classPop_MG_6["CLASSE POP"], classPop_MG_7["CLASSE POP"]]

colors = ['yellowgreen', 'gold', 'purple', 'lightcoral', 'orange', 'gray', 'lightskyblue']
print('BRA:', classSizes)
print('MG:', classMGSizes)

plt.pie(classSizes, labels=classLabel, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Distribuição população brasileira')
#plt.axis('equal')
plt.show()
plt.pie(classMGSizes, labels=classLabel, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Distribuição população mineira')
plt.show()
```
